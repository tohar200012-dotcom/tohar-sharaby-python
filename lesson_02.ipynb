### Part 2: Checking Variable Types
Use the type() function to check what type a variable is:  
Code cell:

```python
# Let's check different types
age = 15
height = 1.75
name = "David"
is_student = True

print(type(age))           # <class 'int'>
print(type(height))        # <class 'float'>
print(type(name))          # <class 'str'>
print(type(is_student))    # <class 'bool'>
```

---
### Part 3: Type Conversion
Sometimes we need to convert between types:  
Converting to Integer

```python
text = "25"
number = int(text)
print(number + 5)  # Result: 30
```

Converting to Float


```python
text = "3.14"
number = float(text)
print(number * 2)  # Result: 6.28
```

Converting to String


```python
age = 15
text = str(age)
print("I am " + text + " years old")
```

Important! You can't convert text that isn't a number:


```python
int("hello")  # This will cause an ERROR!
```

### Challenge for Advanced Students!
Challenge 1: Shopping Calculator  
Write a program that:  

1. Asks for the price of 3 different items  
2. Calculates the total  
3. If the user has a 15% discount coupon, calculate the discounted price  
4. Add 8% tax to the final price  
5. Show all the calculations step by step  

Code cell:  

```python
# Get prices
item1 = float(input("Price of first item: $"))
item2 = float(input("Price of second item: $"))
item3 = float(input("Price of third item: $"))

# Calculate subtotal
subtotal = item1 + item2 + item3

# Ask about discount
has_discount = input("Do you have a discount coupon? (yes/no): ").strip().lower()

# Calculate discount if applicable
if has_discount == "yes":
    discount = subtotal * 0.15
else:
    discount = 0.0

# Price after discount
price_after_discount = subtotal - discount

# Calculate tax (8%)
tax = price_after_discount * 0.08

# Calculate final price
final_price = price_after_discount + tax

# Print detailed receipt (rounded to 2 decimals)
print("\n--- Receipt ---")
print(f"Item 1: ${item1:.2f}")
print(f"Item 2: ${item2:.2f}")
print(f"Item 3: ${item3:.2f}")
print(f"Subtotal: ${subtotal:.2f}")
print(f"Discount: -${discount:.2f}")
print(f"Price after discount: ${price_after_discount:.2f}")
print(f"Tax (8%): +${tax:.2f}")
print(f"Final price: ${final_price:.2f}")
print("--- Thank you! ---")
```
